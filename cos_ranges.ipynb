## Теперь отфильтруем только то, что у нас добавлено в корзину

```python
data_cart = data[data.event_type == "cart"]
```

```python
all_peoples = np.unique(data_cart.user_id)
```

## Формула из презентации

```python
def cosin(a, b):
    """
    Функция считает cos расстояние, по формуле из презентации
    """
    aLength = np.linalg.norm(a)
    bLength = np.linalg.norm(b)

    return np.dot(a, b) / (aLength * bLength)
```

```python
cos_ranges = []
```

```python
our_user = all_peoples[42]  # Выберем рандомно 42 пользователя и для него найдем "соседей"
```

```python
our_user
```

## Создадим список из всех наших продуктов в магазине

```python
all_products = np.unique(data_cart.product_id).tolist()
```

```python
all_products
```

## Теперь найдем продукты нашего пользователя и добавим их количество

```python
products_our_user = data_cart[data_cart.user_id == our_user].product_id.to_list()
```

```python
products_our_user
```

```python
vector1 = [0]*len(all_products)
for x in products_our_user:
    a = all_products.index(x)
    vector1[a] += 1
```

```python
start_time = datetime.now()
```

## Теперь будем сравнивать 2 пользователей, нашего и каждого с 1 по 1000

```python
for i in range(1000):    
    next_user = all_peoples[i]
    products = data_cart[data_cart.user_id == next_user].product_id.to_list()
    
    vector2 = [0]*len(all_products)
    for x in products:
        a = all_products.index(x)
        vector2[a] += 1

    cos_ranges.append(cosin(vector1, vector2))
    
```

```python
end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```
